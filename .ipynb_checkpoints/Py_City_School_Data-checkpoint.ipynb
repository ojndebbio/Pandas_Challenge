# PyCity Schools Analysis

Trend 1: Charter schools out performed the district schools. District schools had a poor performance as it shows in the charts below.

Trebd 2: The average reading score for all schools was better than the average score in maths.(Charter & District schools inclusive).

Trend 3: Charter schools with less budget per student out performed District schools with more budget per student.The high budget for district schools did not translate to good performance in academics. 

#There's no mention of other features that are assigned to quality of performances in both school types.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

csv_path_school = "resources/schools_complete.csv"
csv_path_student = "resources/students_complete.csv"

school_df = pd.read_csv(csv_path_school)
student_df = pd.read_csv(csv_path_student)
```

# District_Summary

```python
# Total Schools
#school_df['name'].count()

# Total Students
#student_df['name'].count()

#Total Budget   
#school_df['budget'].sum()

#Average Math Score   
#round(student_df['math_score'].mean(),2)

#Average Reading Score
#round(student_df['reading_score'].mean(),2)

# % Passing Math
math_score_pass = student_df[student_df.math_score > 70]
m = (math_score_pass['math_score'].count()/student_df['student_name'].count())*100

# % Passing Reading
reading_score_pass = student_df[student_df.reading_score > 70]
r = (reading_score_pass['reading_score'].count()/student_df['student_name'].count())*100

# % Overall Passing rate


d = {'Total Schools': [school_df['school_name'].count()], 
     'Total Students': [student_df['student_name'].count()], 
     'Total Budget':[school_df['budget'].sum()], 
     'Average Math score':[round(student_df['math_score'].mean(),2)], 
     'Average Reading score':[round(student_df['reading_score'].mean(),2)],
     '% Passing Math':[round(m,2)],
     '% Passing Reading':[round(r,2)],
     '% Overall Passing Rate':[round((m+r)/2,2)]
    }

school_summary = pd.DataFrame(d)
school_summary = school_summary.loc[:,['Total Schools','Total Students','Total Budget','Average Math score',
                                       'Average Reading score','% Passing Math','% Passing Reading', '% Overall Passing Rate']]

school_summary['Total Budget'] = school_summary['Total Budget'].map("${:,.0f}".format)
school_summary['Total Students'] = school_summary['Total Students'].map("{:,.0f}".format)
school_summary['% Passing Math'] = school_summary['% Passing Math'].map("{:,}%".format)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map("{:,}%".format)
school_summary['% Overall Passing Rate'] = school_summary['% Overall Passing Rate'].map("{:,}%".format)

school_summary
```

```python
d1 = school_df.loc[:,['school_name', 'type', 'size', 'budget']]
sch = pd.DataFrame(d1)

sch.rename(columns={'type': 'School Type',
                    'size': 'Total Students',
                    'budget': 'Total School Budget',
                    
                   }, inplace=True)

sch['Per Student Budget'] = sch['Total School Budget']/sch['Total Students']

#sch['Per Student Budget'] = sch['Per Student Budget'].map("${:,}".format)
sch
```

```python
d2 = student_df.groupby(['school_name'])
stu = pd.DataFrame(round(d2.mean(),2))   
stu = stu.loc[:,['reading_score','math_score']]
stu
```

```python
reading_passout = student_df.copy()
reading_passout = reading_passout.loc[:,['school_name','reading_score']]
reading_passout['reading_score'] = reading_passout.loc[reading_passout['reading_score']>70]
reading_passout = reading_passout.groupby('school_name')
reading_passout = pd.DataFrame(reading_passout.count())
reading_passout
```

```python
math_passout = student_df.copy()
math_passout = math_passout.loc[:,['school_name','math_score']]
math_passout['math_score'] = math_passout.loc[math_passout['math_score']>70]
math_passout = math_passout.groupby('school_name')
math_passout = pd.DataFrame(math_passout.count())
math_passout
```

```python
passout = reading_passout.join(math_passout)
passout = pd.DataFrame(passout)
passout.rename(columns={'reading_score':'Reading Passout', 'math_score':'Math Passout'}, inplace=True)
passout
```

# School_Summary

```python
xyz = sch.join(stu, on='school_name')
final_frame = pd.DataFrame(xyz)

final_frame = final_frame.join(passout, on='school_name')
final_frame = pd.DataFrame(final_frame)
final_frame['% Passing Reading'] = round(final_frame['Reading Passout']/final_frame['Total Students']*100,2)
final_frame['% Passing Math'] = round(final_frame['Math Passout']/final_frame['Total Students']*100,2)
final_frame['Overall Passing Rate'] = (final_frame['% Passing Reading']+final_frame['% Passing Math'])/2
final_frame.rename(columns={'reading_score':'Average Reading Score',
                            'math_score':'Average Math Score'
                            }, inplace=True)
final_frame
```

## Performing Top Schools (By Passing Rate)

## Performing Bottom Schools (By Passing Rate)

```python
final_frame.nsmallest(5, 'Overall Passing Rate')
```

## Reading Score (Using Grades)

```python
grade_arrange_reading = student_df.copy()
grade_arrange_reading.head()
grade_arrange_reading = pd.pivot_table(grade_arrange_reading, index=['school_name'], columns='grade', values='reading_score').reset_index()
grade_arrange_reading
```

# Math Scores (Using Grades)

```python
grade_arrange_math = student_df.copy()
grade_arrange_math.head()
grade_arrange_math = pd.pivot_table(grade_arrange_math, index=['school_name'], columns='grade', values='math_score').reset_index()
grade_arrange_math
```

# Scores by School Spending/Budget

```python
school_spending = final_frame.copy()
lable_spendind = np.array(['577.99 to 591.5', '591.5 to 628.0', '628.0 to 641.5', '641.5 to 655.0'])
school_spending['Per Student Budget'] = pd.qcut(school_spending['Per Student Budget'], 4, labels=lable_spendind )
school_spending.groupby(['Per Student Budget'])['Average Reading Score', 'Average Math Score', '% Passing Reading', '% Passing Math',
                                              'Overall Passing Rate'].mean()
```

# Scores by School Size

```python
school_size_arrange = final_frame.copy()

lable_size = np.array(['Small','Medium', 'Large'])
school_size_arrange['Total Students'] = pd.qcut(school_size_arrange['Total Students'], 3, labels=lable_size )

school_size_arrange.groupby(['Total Students'])['Average Reading Score', 'Average Math Score', '% Passing Reading', '% Passing Math',
                                            'Overall Passing Rate'].mean()
```
